```python
# df.drop(columns='date',inplace=True)  
# df.drop('date', axis=1,inplace=True)
df = df.drop('date', axis=1)
```

```python
df = df.drop('country', axis=1)
```

```python
# for object or string
df['street'].mode()[0]   # most frequent value
```

```python
# df['street'].fillna(df['street'].mode()[0])
df['street'] = df['street'].fillna(df['street'].mode()[0])
```

```python
def fillnaobjmode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
columns = ['street','city','statezip']
fillnaobjmode(columns)
```

```python
# for float values
def fillnafloatmean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())
columns = ['price','sqft_living','sqft_lot','condition','sqft_above','sqft_basement']
fillnafloatmean(columns)
```

```python
# for int values
def fillnaintmode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
columns = ['bedrooms','bathrooms','waterfront','floors','view','yr_built','yr_renovated']
fillnaintmode(columns)
```

```python
df['price'] = df['price'].astype('int64')
```

```python

def changefloat(cols):
    for i in cols:
        df[i] = df[i].astype('int64')
columns = ['price','sqft_living','sqft_lot','condition',
           'sqft_above','sqft_basement','bedrooms','bathrooms','waterfront',
           'floors','view','yr_built','yr_renovated']
changefloat(columns)
    
```

```python
data = pd.DataFrame({'city':df['city'].unique()})
data_LE = LabelEncoder()
data_LE.fit(np.ravel(data))
```

```python
def changeobjtoint(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        # data_LE.fit(data)
        df[i] = data_LE.transform(df[i])
columns = ['street','city','statezip']
changeobjtoint(columns)
    
```
