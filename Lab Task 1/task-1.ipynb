```python
word = "python"
attempt = 6
guessed = ["_"] * len(word)

while attempt > 0 and "_" in guessed:
    print("Word:", guessed)
    guess = input("Enter a letter: ")

    if guess == "p":
        print("correct")
        guessed[0] = "p"
    elif guess == "y":
        print("correct")
        guessed[1] = "y"
    elif guess == "t":
        print("correct")
        guessed[2] = "t"
    elif guess == "h":
        print("correct")
        guessed[3] = "h"
    elif guess == "o":
        print("correct")
        guessed[4] = "o"
    elif guess == "n":
        print("correct")
        guessed[5] = "n"
    else:
        attempt -= 1
        print("wrong, attempts left:", attempt)

        
        if attempt == 5:
            print("---------|")
            print("|        O")
        elif attempt == 4:
            print("---------|")
            print("|        O")
            print("|        |")
        elif attempt == 3:
            print("---------|")
            print("|        O")
            print("|       /|")
        elif attempt == 2:
            print("---------|")
            print("|        O")
            print("|       /|\\")
        elif attempt == 1:
            print("---------|")
            print("|        O")
            print("|       /|\\")
            print("|       /")
        elif attempt == 0:
            print("---------|")
            print("|        O")
            print("|       /|\\")
            print("|       / \\")
            print("Game Over!")

if "_" not in guessed:
    print("You won! The word was:", word)
else:
    print("You lost! The word was:", word)
```

```python
import random

words = ["python", "banana", "orange", "lahore"]
word = random.choice(words)  
attempt = 6
guessed = ["_"] * len(word)

while attempt > 0 and "_" in guessed:
    print("Word:", guessed)
    guess = input("Enter a letter: ").lower()

    if guess in word:   
        print("correct")
        for i in range(len(word)):
            if word[i] == guess:
                guessed[i] = guess
    else:
        attempt -= 1
        print("wrong, attempts left:", attempt)

     
        if attempt == 5:
            print("---------|")
            print("|        O")
        elif attempt == 4:
            print("---------|")
            print("|        O")
            print("|        |")
        elif attempt == 3:
            print("---------|")
            print("|        O")
            print("|       /|")
        elif attempt == 2:
            print("---------|")
            print("|        O")
            print("|       /|\\")
        elif attempt == 1:
            print("---------|")
            print("|        O")
            print("|       /|\\")
            print("|       /")
        elif attempt == 0:
            print("---------|")
            print("|        O")
            print("|       /|\\")
            print("|       / \\")
            print("Game Over!")

if "_" not in guessed:
    print("You won! The word was:", word)
else:
    print("You lost! The word was:", word)
```
