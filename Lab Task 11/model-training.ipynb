```python
# df.drop(columns='date',inplace=True)  
# df.drop('date', axis=1,inplace=True)
df = df.drop('date', axis=1)
```

```python
df = df.drop('country', axis=1)
```

```python
# for object or string
df['street'].mode()[0]   # most frequent value
```

```python
# df['street'].fillna(df['street'].mode()[0])
df['street'] = df['street'].fillna(df['street'].mode()[0])
```

```python
def fillnaobjmode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
columns = ['street','city','statezip']
fillnaobjmode(columns)
```

```python
# for float values
def fillnafloatmean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())
columns = ['price','sqft_living','sqft_lot','condition','sqft_above','sqft_basement']
fillnafloatmean(columns)
```

```python
# for int values
def fillnaintmode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
columns = ['bedrooms','bathrooms','waterfront','floors','view','yr_built','yr_renovated']
fillnaintmode(columns)
```

```python
df = pd.read_csv("preprocessed.csv")
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('price',axis=1) # with resultant column
y = df['price']    # with resultant column
```

```python
x.head(1)
```

```python
y.head(1)
```

```python
train_X, test_X, train_y, test_y = train_test_split(x,y,test_size=0.3,shuffle=False) # 70 percent training and 30 percent testing
```

```python
train_X
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

# apply classifier/ model training

```python
from sklearn.svm import SVC
modelsvc = SVC()
modelsvc.fit(train_X, train_y)
print(modelsvc)
```

```python
import pickle
pickle.dump(modelsvc,open('modelsvc.pkl','wb'))
```

```python
#loading model
modelsvc = pickle.load(open('modelsvc.pkl','rb'))
```

```python
modelsvcpred = modelsvc.predict(test_X)
modelsvcpred
```

```python
from sklearn.metrics import accuracy_score
modelsvcaccu = accuracy_score(test_y,modelsvcpred)
print(round(modelsvcaccu,3)*100,'%')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import numpy as np

RF = RandomForestClassifier()


RF.fit(train_X, train_y)


Y_rpred = RF.predict(test_X)


r_accuracy = metrics.accuracy_score(test_y, Y_rpred)
print("Accuracy: {:.2f}%".format(r_accuracy * 100))



r_precision = metrics.precision_score(test_y, Y_rpred, average='weighted', zero_division=0)
print("Precision: %f" % r_precision)

r_recall = metrics.recall_score(test_y, Y_rpred, average='weighted', zero_division=0)
print("Recall: %f" % r_recall)


r_f1 = metrics.f1_score(test_y, Y_rpred, average='weighted', zero_division=0)
print("F1 Score: %f" % r_f1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

Dtree = DecisionTreeClassifier(random_state=42)


Dtree.fit(train_X,train_y)


Y_dpred = Dtree.predict(test_X)


d_accuracy = accuracy_score(test_y, Y_dpred)
d_precision = precision_score(test_y, Y_dpred, average='weighted', zero_division=0)
d_recall = recall_score(test_y, Y_dpred, average='weighted', zero_division=0)
d_f1 = f1_score(test_y, Y_dpred, average='weighted', zero_division=0)


print('Decision Tree Classifier Metrics:')
print(f'Accuracy:  {d_accuracy*100:.2f}%')
print(f'Precision: {d_precision:.4f}')
print(f'Recall:    {d_recall:.4f}')
print(f'F1 Score:  {d_f1:.4f}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

KNN = KNeighborsClassifier()


KNN.fit(train_X,train_y)


Y_kpred = KNN.predict(test_X)


k_accuracy = accuracy_score(test_y, Y_kpred)
k_precision = precision_score(test_y, Y_kpred, average='weighted', zero_division=0)
k_recall = recall_score(test_y, Y_kpred, average='weighted', zero_division=0)
k_f1 = f1_score(test_y, Y_kpred, average='weighted', zero_division=0)


print('KNN Classifier Metrics:')
print(f'Accuracy:  {k_accuracy*100:.2f}%')
print(f'Precision: {k_precision:.4f}')
print(f'Recall:    {k_recall:.4f}')
print(f'F1 Score:  {k_f1:.4f}')
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np


xgb = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=5,
    random_state=42
)

xgb.fit(train_X, train_y)
Y_pred = xgb.predict(test_X)


mse = mean_squared_error(test_y, Y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test_y, Y_pred)
r2 = r2_score(test_y, Y_pred)

r2_percent = r2 * 100

target_range = test_y.max() - test_y.min()

rmse_percent = (1 - rmse / target_range) * 100
mae_percent = (1 - mae / target_range) * 100
mse_percent = (1 - mse / (target_range ** 2)) * 100


print(f"Accuracy:  {rmse_percent:.2f}%")
print(f"Precision: {mae_percent:.2f}%")
print(f"Recall:    {mse_percent:.2f}%")
print(f"F1 Score:  {r2_percent:.2f}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np


classifiers = ['Random Forest', 'Decision Tree', 'KNN', 'XGBoost']

accuracy = [r_accuracy, d_accuracy, k_accuracy, rmse_percent]
precision = [r_precision, d_precision, k_precision, mae_percent]
recall = [r_recall, d_recall, k_recall, mse_percent]
f1 = [r_f1, d_f1, k_f1, r2_percent]

x = np.arange(len(classifiers))
width = 0.2

plt.figure(figsize=(12,7))

plt.bar(x - 1.5*width, accuracy, width, label='Accuracy', color='blue')
plt.bar(x - 0.5*width, precision, width, label='Precision', color='green')
plt.bar(x + 0.5*width, recall, width, label='Recall', color='red')
plt.bar(x + 1.5*width, f1, width, label='F1 Score', color='purple')

plt.xticks(x, classifiers)
plt.ylim(0, 100) 
plt.xlabel("Model")
plt.ylabel("Score (%)")
plt.title("Top 3 Classifiers + XGBoost Metrics Comparison")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
