```python
from sklearn.preprocessing import LabelEncoder
data = pd.DataFrame({'Make':df['Make'].unique()})
data_LE = LabelEncoder()
data_LE.fit(np.ravel(data))
```

```python
def changeobjtoint(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        # data_LE.fit(data)
        df[i] = data_LE.transform(df[i])
columns = ["Make", "Model", "FuelType", "Transmission", "Location", "Color", "Owner", "SellerType", "Drivetrain"]
changeobjtoint(columns)
    
```
